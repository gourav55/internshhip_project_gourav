```python
data_x=loanset.loc[:,loanset.columns!='Personal Loan']
data_y=loanset[['Personal Loan']]
```

# Transformation of feature variables

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method="yeo-johnson",standardize=False)
pt.fit(data_x["Income"].values.reshape(-1,1))
temp=pt.transform(data_x["Income"].values.reshape(-1,1))
sns.distplot(temp);
```

```python
pt=PowerTransformer(method="yeo-johnson",standardize=False)
pt.fit(data_x["CCAvg"].values.reshape(-1,1))
temp=pt.transform(data_x["CCAvg"].values.reshape(-1,1))
sns.distplot(temp);
```

```python
data_x["Mortgage_Int"]=pd.cut(data_x["Mortgage"],
                             bins=[0,100,200,300,400,500,600,700],
                             labels=[0,1,2,3,4,5,6],
                             include_lowest=True)
data_x.drop("Mortgage",axis=1,inplace=True)
```

```python
data_x.head()
```

```python
tempDF=pd.DataFrame(loanset["Personal Loan"].value_counts()).reset_index()
tempDF.columns=["Labels","Personal Loan"]
fig1,ax1=plt.subplots(figsize=(10,8))
explode=(0,0.15)
ax1.pie(tempDF["Personal Loan"],explode=explode,autopct='%1.1f%%',shadow=True,startangle=70)
ax1.axis('equal')
plt.title("Personal Loan Pencentage")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xTrain,xTest,yTrain,yTest=train_test_split(data_x,data_y,train_size=0.6,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
Log_model=LogisticRegression()
```

```python
Log_model.fit(xTrain,yTrain)
```

```python
from sklearn.metrics import accuracy_score
pred=Log_model.predict(xTrain)
accuracy_score(yTrain,pred)
```

```python
pred_ontest=Log_model.predict(xTest)
accuracy_score(yTest,pred_ontest)
```

```python
from sklearn.metrics import confusion_matrix
confusion=confusion_matrix(yTest,pred_ontest)
confusion
```

```python
TP=confusion[1,1]          #true positive
TN=confusion[0,0]          #true negative
FP=confusion[0,1]          #false positive
FN=confusion[1,0]          #false negative
```

```python
#recall or sensitivity or true_positive_rate
recall=TP/(TP+FN)
recall
```

```python
#specificity or (1-false_positive_rate)
specificity=TN/(TN+FP)
specificity
```

```python
#precision
precision=TP/(TP+FP)
precision
```

```python
#positive predictive value
positive_predictive_value=TP/(TP+FP)
positive_predictive_value
```

```python
#Negative predictive value
negative_predictive_value=TN/(TN+FN)
negative_predictive_value
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
auc_score=roc_auc_score(yTest,pred_ontest)
fpr,tpr,thresholds=roc_curve(yTest,Log_model.predict_proba(xTest)[:,1])
```

```python
plt.figure()
plt.plot(fpr,tpr,label="Logistic Model 1(area=%0.2f)"%auc_score)
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title("Receiver Operating Characteristics")
plt.show()
```
